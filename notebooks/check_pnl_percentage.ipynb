```python
import pandas as pd
from pathlib import Path
```

```python
live_trades = Path("../logs/live_trades_BTC-USD_20250825_234338_mock.csv")
```

```python
df = pd.read_csv(live_trades)
df.head()
```

```python
df.shape
```

```python
for idx, row in df.iterrows():
    print(row['timestamp'])
```

```python
initial_amt = 10000
pnls = []
pnls_pct = []
trade_type = None
for idx, row in df.iterrows():
    if idx%2 == 0:
        #taking trade
        trade_type = str(row['action'])
        entry_price = row['price']
        print(f" {trade_type}, entry_price : {entry_price}")
    else:
        #exit the trade, cal pnl and balance
        exit_trade_type = str(row['action'])
        if exit_trade_type != 'EXIT':
            print(f"ERROR {idx}")
        exit_price = row['price']
        print(f"EXIT, trade_type: {trade_type} , exit_price : {exit_price}")
        if trade_type == 'SELL':
            pnl = entry_price - exit_price
            pnl_pct = (entry_price - exit_price)/entry_price *100
        else:
            pnl = exit_price - entry_price
            pnl_pct = (exit_price - entry_price)/entry_price *100
        pnls.append(pnl)
        pnls_pct.append(pnl_pct)
        initial_amt += initial_amt*pnl_pct/100
        
        
        
```
